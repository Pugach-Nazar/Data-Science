1. Здійснити загальні операції з використанням TensorFlow (створити набір даних (розмір за власним рішенням) у вигляді тензорів, провести CRUD операції над даними (скаляр, вектор, матриця)).



- Створення тензорів

```python
import tensorflow as tf
import numpy

scalar = tf.constant(123)

vector = tf.constant([1, 2, 3])

matrix = tf.random.Generator.from_seed(42)
matrix = matrix.normal(shape=(3, 2))
```

- Читання тензорів

```python

print("скаляр:", scalar.numpy())

print("вектор:", vector.numpy())

print("матриця:\n", matrix.numpy())
```

- Оновлення тензорів

```python
updated_scalar = scalar + 10
print("Оновлений скаляр:", updated_scalar.numpy())

updated_vector = vector + 1
print("Оновлений вектор:", updated_vector.numpy())

updated_matrix = matrix * 2
print("Оновлена матриця:\n", updated_matrix.numpy())
```

- Видалення тензорів

```python
del scalar, vector, matrix, updated_scalar, updated_vector, updated_matrix
```

2. Згенерувати набір даних для регресії та 
класифікації з використанням TensorFlow



- Генерація набору даних для регресії

```python
n_samples = 100 
X_regression = tf.random.normal(shape=(n_samples, 1), )
noise = tf.random.normal(shape=(n_samples, 1), stddev=0.5)
y_regression = 3 * X_regression + 7 + noise

print("Вхідні дані:", X_regression[:5].numpy())
print("Цільові дані:", y_regression[:5].numpy())
```

- Генераця набору даних для класифікації

```python
n_samples_class = 100
class_0 = tf.random.normal(shape=(n_samples_class, 2), mean=3.0, stddev=1.0)
class_1 = tf.random.normal(shape=(n_samples_class, 2), mean=6.0, stddev=1.0)

X_classification = tf.concat([class_0, class_1], axis=0)
y_classification = tf.concat([tf.zeros((n_samples_class, 1)), tf.ones((n_samples_class, 1))], axis=0)

print("Вхідні дані (класифікація):", X_classification[:5].numpy())
print("Цільові дані (класифікація):", y_classification[:5].numpy())
```

3. Здійснити графічний аналіз наборів даних за допомогою Matplotlib

- Візуалізація набору даних для регресії

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X_regression.numpy(), y_regression.numpy(), color='blue', alpha=0.5, label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Набір даних для регресії")
plt.legend()
plt.show()
```

- Візуалізація набору даних для класифікації

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_classification[:n_samples_class, 0], X_classification[:n_samples_class, 1], 
            color='red', alpha=0.5, label="Class 0")
plt.scatter(X_classification[n_samples_class:, 0], X_classification[n_samples_class:, 1], 
            color='green', alpha=0.5, label="Class 1")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Набір даних для класифікації")
plt.legend()
plt.show()
```
